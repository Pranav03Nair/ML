### Using MatplotLib

LR for a predicting House Prices with given data set -
 
House Size - [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]

House Price - [300, 450, 500, 550, 600, 650, 700, 750, 800]

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]).reshape(-1, 1)
Y = np.array([300, 450, 500, 550, 600, 650, 700, 750, 800])
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
x_range = np.linspace(900,5200,100).reshape(-1,1)
predicted_prices = model.predict(x_range)
```

```python
# Plot DataPoints
plt.scatter(X, Y, color='blue',label='Data Points')

# Plot linear Regression
plt.scatter(x_range, predicted_prices, color='red',label='LR Line')

#Setting Labels
plt.xlabel('HouseSize')
plt.ylabel('Price')
plt.title('LR: House Prices')

#Legend
plt.legend()

#show plot
plt.show()
```
